# Import Pandas

```python
import pandas as pd
```

```python
# Creating the self dataset and perform some Pandas functions on it
dataset = {'Student': ['Ronit','Yuvraj','Hansal','Zara'],
          'Admission': [2021,2025,2022,2021]}
# Dataframe is used to access the dataset and convert it into structured format
college = pd.DataFrame(dataset)
print(college)
```

```python
# Check the version of Pandas
pd.__version__
```

```python
# Series function are used to bind the data in single rows format
# We can also change the Index format of the dataset here
a = [10,20,30,40,50]
ser = pd.Series(a)
print(ser)
```

```python
# We can bind the data with another data into it
marks = {"Ronit": 70, "Zara": 66,"Yuvraj":70}
c_marks = pd.Series(marks)
print(c_marks)
```

```python
# To fetch the value of the data from data in key
print(c_marks["Ronit"])
```

```python
#  To fetch the value of data through value
print(c_marks.iloc[1])
```

```python
# Sliceing the dataset
print(a[1:])
```

```python
# Create a dataset from your choice and print it
data = {"Product": ['Tea', 'Coffee', 'Puff', 'Dabeli', 'Water', 'Parle-G'], "Price": [20,30,40,25,10,30]}
Items = pd.DataFrame(data)
print(Items)
```

```python
# From the dataset try to slice the Items
print(Items.loc[3:])
```

```python
# Import the CSV file into our pratice file
Stock = pd.read_csv("Warehouse_Stock.csv")
print(Stock)
```

```python
# Print the Total Sales in existing CSV file by calculating the Stock Sold and PerPerUnit
Stock['TotalSales'] = (Stock['Stock Sold'] * Stock['PricePerUnit'])
print(Stock)
```

```python
# Print the First 4 value from CSV file
Stock.head(4)
```

```python
# Print the last 3 values from the csv file
Stock.tail(3)
```

```python
# Print any one value from Stock in Ascending order
Stock.sort_values(['Stock Available'])
```

```python
# Perform the Mean, Median and Mode operation's on the CSV file
print(Stock['TotalSales'].mean())
print(Stock['TotalSales'].median())
print(Stock['TotalSales'].mode())
```
