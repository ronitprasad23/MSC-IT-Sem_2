# Import Numpy

```python
import numpy as np
```

```python
# Create an array
a = np.array([1,3,7,9,10,18])
```

```python
# Perform In-build library on given array

print(a)
print("shape", a.shape)
print("Dimension", a.ndim)
print("Size", a.size)
print("Data Type", a.dtype)
print("Item Size", a.itemsize)
```

# Two Dimenstional Array

```python
# Taking Two Dimentional Array
al = np.array([[1,2,3], [6,9,6]])
print(al)
a2 = np.array([[3,9,9], [8,4,8]])
print(a2)
#Addititon
Add = np.add(al , a2)
print("Addition", Add)
# Substraction
Sub = np.subtract(al, a2)
print("Substraction", Sub)
# Divide
Divide = np.divide(al, a2)
print("Divide", Divide)
# Multiply
Mult = np.multiply(al, a2)
print("Multiply", Mult)
# Mod
Mod = np.mod(al, a2)
print("Mod", Mod)
# Remainder
Rem = np.remainder(al, a2)
print("Remainder", Rem)
# power
Pow = np.power(al, a2)
print("Power", Pow)
# Absolute
Abso = np.absolute(al, a2)
print("Absolute", Abso)
# Sum of a single array
print("Sum", np.sum(al))
print("Sqrt", np.sqrt(al))
```

```python
# Metrix Multiplication
b1 = np.array([[2,7,9], [8,6,5]])
b2 = np.array([[3,8,4], [1,6,8]])
mul = np.dot(b1, b2.T)
print("Multiplication", mul)
```
